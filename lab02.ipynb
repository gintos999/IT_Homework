# Задача 3

```python
df_games = pd.read_csv('games001.csv', sep = ';')
df_games
```

```python
df_rates = pd.read_csv('rates001.csv', sep = ';')
df_rates
```

```python
marks = []
for i in df_games['id']:
    marks.append(df_rates[(df_rates['id'] == i)]['mark'].mean())
data = {'id':df_games['id'], 'name':df_games['name'], 'mark':marks, 'company':df_games['company']}
df_res = pd.DataFrame(data).sort_values(by = 'mark', ascending = False)
display(df_res)

for i in range(3):
    print(df_res.iloc[i][1], f'{df_res.iloc[i][2]:.3f}')
```

```python
df = df_res[(df_res['mark'] > 8)]
display(df)
print(df['company'].value_counts(ascending=False).index[0], df['company'].value_counts(ascending=False)[0])
```
